```python
import pandas as pd
```

```python
df = pd.read_csv('SeattleWeather.csv', index_col=["Unnamed: 0", "Unnamed: 0.1"])
```

```python
df.head()
```

```python
'''
Winter: Dec - Feb
Spring: March - May
Summer June-Aug
Fall Sept-Nov
'''
```

```python
df.rename(columns={'Location': 'Precinct'}, inplace=True)
```

```python
df['Precinct'] = df['Precinct'].str.upper()
```

```python
df.head()
```

```python
df.loc[(df.Month == 12) | (df.Month == 1) | (df.Month == 2), 'Season'] = 'WINTER'
df.loc[(df.Month == 3) | (df.Month == 4) | (df.Month == 5), 'Season'] = 'SPRING'
df.loc[(df.Month == 6) | (df.Month == 7) | (df.Month == 8), 'Season'] = 'SUMMER'
df.loc[(df.Month == 9) | (df.Month == 10) | (df.Month == 11), 'Season'] = 'FALL'
```

```python
df.head()
```

```python
def calcMean(df, col, val, target):
    mean = df[df[col]==val][target].mean()
    print("The mean temperatures for the season of " + val + " is: ")
    print("\n")
    print(mean)
    return mean

def calcSTD(df, col, val, target):
    std = df[df[col]==val][target].std()
    print("The standard deviation in the temperatures for the season of " + val + " is: ")
    print("\n")
    print(std)
    return std
```
